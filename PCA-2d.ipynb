```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = PCA(n_components=2)

X_r = pca.fit(X).transform(X) #reduced version of x
%matplotlib inline
#plt.scatter(X_r[:,0],X_r[:,1])  
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], iris.target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)


    
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = PCA(n_components=3)

X_r = pca.fit(X).transform(X) #reduced version of x
%matplotlib inline
# plt.scatter(X_r[:,0],X_r[:,1])  
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], iris.target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], X_r[y == i, 2], color=color, alpha=.8,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
```
